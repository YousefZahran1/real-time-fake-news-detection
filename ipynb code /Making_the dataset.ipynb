```python

import requests
import pandas as pd

API_KEYS = [
    "3c26db8b9d234aa08d05dceec408609b",
    "1f19cef3d71547879eca11f3498530cf",
    "4896eff9b572410cbd4802cf2aff3abd",
    "81c0e92474a9423abd6218e45ad48e5a",
    # Add more keys here if you get them
]
sources = "bbc-news,cnn,the-new-york-times,reuters,associated-press"
page_size = 100

all_articles = []
key_idx = 0

# You can add keywords or date ranges for more variety
keywords = ["", "politics", "economy", "science", "health", "war", "technology"]

for keyword in keywords:
    api_key = API_KEYS[key_idx % len(API_KEYS)]
    key_idx += 1
    print(f"Using API key {api_key[:4]}... for keyword: {keyword}")
    url = (
        f'https://newsapi.org/v2/everything?'
        f'sources={sources}&'
        f'pageSize={page_size}&'
        f'apiKey={api_key}'
    )
    if keyword:
        url += f"&q={keyword}"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error: {response.status_code}, {response.text}")
        continue
    articles = response.json().get('articles', [])
    for art in articles:
        text = art.get('content') or art.get('description') or art.get('title', '')
        if text and len(text) > 100:
            all_articles.append({
                'title': art.get('title', ''),
                'text': text,
                'description': art.get('description', ''),
                'url': art.get('url', ''),
                'publishedAt': art.get('publishedAt', ''),
                'source': art.get('source', {}).get('name', ''),
                'label': 1
            })

df = pd.DataFrame(all_articles)
df.to_csv('/content/drive/MyDrive/FYP Datasets/real_news_newsapi_manykeys.csv', index=False)
print(f"Saved {len(df)} real news articles using multiple API keys and keywords.")
'''
```

```python
import pandas as pd
import os

# 1. Paths (update if different)
base = "/content/drive/MyDrive/FYP Datasets/"

# === Fake-and-Real-News Dataset ===
kaggle_fake = pd.read_csv(base + "fake-and-real-news-dataset/Fake.csv")
kaggle_real = pd.read_csv(base + "fake-and-real-news-dataset/True.csv")
# Add source, label, url if missing
for df, src, label in [(kaggle_fake, "Kaggle", 0), (kaggle_real, "Kaggle", 1)]:
    df['source'] = src
    df['label'] = label
    if 'url' not in df.columns:
        df['url'] = ""
    if 'title' not in df.columns:
        df['title'] = df['text'].str[:50]  # fallback: first 50 chars

# === FakeNewsNet Dataset (PolitiFact + GossipCop, real and fake) ===
fnn_paths = [
    ("FakeNewsNet-dataset/politifact_fake_cleaned.csv", 0, "PolitiFact"),
    ("FakeNewsNet-dataset/politifact_real_cleaned.csv", 1, "PolitiFact"),
    ("FakeNewsNet-dataset/gossipcop_fake_cleaned.csv", 0, "GossipCop"),
    ("FakeNewsNet-dataset/gossipcop_real_cleaned.csv", 1, "GossipCop"),
]
fnn_dfs = []
for path, label, src in fnn_paths:
    df = pd.read_csv(base + path)
    df['source'] = src
    df['label'] = label
    # Try to pick best column for text
    if 'clean_text' in df.columns:
        df['text'] = df['clean_text']
    elif 'statement' in df.columns:
        df['text'] = df['statement']
    if 'title' not in df.columns:
        df['title'] = df['text'].str[:50]
    if 'url' not in df.columns:
        df['url'] = df.get('news_url', "")
    fnn_dfs.append(df)

# === LIAR Dataset (train, valid, test) ===
liar_paths = [
    "liar_dataset/train_cleaned.csv",
    "liar_dataset/valid_cleaned.csv",
    "liar_dataset/test_cleaned.csv"
]
liar_dfs = []
for path in liar_paths:
    df = pd.read_csv(base + path)
    df['source'] = "LIAR"
    # Label: try binary_label or label column (map fake/real)
    if 'binary_label' in df.columns:
        df['label'] = df['binary_label']
    elif 'label' in df.columns:
        df['label'] = df['label'].map({
            "pants-fire": 0, "false": 0, "barely-true": 0,
            "half-true": 1, "mostly-true": 1, "true": 1
        })
    if 'text' not in df.columns:
        df['text'] = df['statement']
    if 'title' not in df.columns:
        df['title'] = df['statement'].str[:50]
    if 'url' not in df.columns:
        df['url'] = ""
    liar_dfs.append(df)

# === Keep Only Desired Columns ===
keep_cols = ['source', 'text', 'label', 'title', 'url']

all_dfs = [
    kaggle_fake[keep_cols],
    kaggle_real[keep_cols],
    *[df[keep_cols] for df in fnn_dfs],
    *[df[keep_cols] for df in liar_dfs],
]
combined = pd.concat(all_dfs, ignore_index=True).drop_duplicates(subset="text").reset_index(drop=True)

# === Save to CSV ===
combined.to_csv(base + "combined_fake_news.csv", index=False)
print("Saved:", combined.shape, base + "combined_fake_news.csv")
```

```python
import pandas as pd

# File paths
path1 = '/content/drive/MyDrive/FYP Datasets/real_news_newsapi_english_langid.csv'
path2 = '/content/drive/MyDrive/FYP Datasets/real_news_rss_english_big.csv'
path3 = '/content/drive/MyDrive/FYP Datasets/combined_fake_news_1.csv'
output_path = '/content/drive/MyDrive/FYP Datasets/combined_fake_news_2.csv'

# Desired columns
cols = ['source', 'text', 'label', 'title', 'url']

# Load each file
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df3 = pd.read_csv(path3)

# Add missing columns with default values where needed
for df, src_name in zip([df1, df2], ["newsapi", "rss"]):
    if 'source' not in df.columns:
        df['source'] = src_name
    if 'label' not in df.columns:
        df['label'] = 1  # Mark all as real news
    if 'title' not in df.columns:
        df['title'] = df['text'].str[:50]
    if 'url' not in df.columns:
        df['url'] = ""

# Reorder columns
df1 = df1[cols]
df2 = df2[cols]
df3 = df3[cols]

# Concatenate all dataframes
df_all = pd.concat([df1, df2, df3], ignore_index=True)

# Remove duplicates based on text
df_all = df_all.drop_duplicates(subset='text').reset_index(drop=True)

# Save the combined file
df_all.to_csv(output_path, index=False)
print(f"Combined dataset saved as {output_path}")
print(f"Total samples: {len(df_all)}")
```
